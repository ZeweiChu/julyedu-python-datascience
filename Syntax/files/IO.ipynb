## concatenate a list of files to be a single file

```python
basedir = "/Users/zeweichu/julyedu/python-data-science/Basics/"
files = ['1-unix.md', '2-intro.md', '2-intro.ppt', '3-Python-basics.md', '4-type-variable-operation-expression.md', '5-list.md', '6-string.md', 'basic.py']
content = ""
for fname in files:
    if fname.endswith(".md"):
        file = basedir + fname
        # print(file)
        try:
            with open(file, "r") as f:
                content += f.read()
        except IOError:
            print("file {} does not exist!".format(file))

print(content)
```
