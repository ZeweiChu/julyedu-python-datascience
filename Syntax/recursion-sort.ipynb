n! = n * (n-1) * (n-2) ... * 1

```python
def factorial(n):
    if n in [1, 0]:
        return 1
    else:
        return n*factorial(n-1)
```

```python
for i in range(10):
    print(factorial(i))
```

```python
def gcd(p, q):
    if q == 0:
        return p
    else:
        return gcd(q, p%q)
```

```python
gcd(5,10)
```

```python
def fibonacci(n):
    if n in [0, 1]:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(5) # 0, 1, 1, 2, 3, 5
```

```python
fibonacci(10)
```

```python
for i in range(10):
    print(fibonacci(i))
```

```python
fibonacci(50)
```

```python
def fibonacci2(n):
    a, b = 0, 1
    for i in range(n):
        a, b = b, a+b
    return a
```

```python
fibonacci2(100)
```

## Sort

```python
arr = [5, 1, 10, 22, -9, -100, 55]
```

```python
sorted(arr)
```

```python
def sort(arr):
    length = len(arr)
    if length in [0, 1]:
        return arr
    else:
        left_arr = sort(arr[:length//2])
        right_arr = sort(arr[length//2:])
        
        new_arr = []
        i = 0
        j = 0
        left_length = len(left_arr)
        right_length = len(right_arr)
        while i < left_length or j < right_length:
            if i >= left_length:
                new_arr.append(right_arr[j])
                j += 1
            elif j >= right_length:
                new_arr.append(left_arr[i])
                i += 1
            else:
                if left_arr[i] < right_arr[j]:
                    new_arr.append(left_arr[i])
                    i += 1
                else:
                    new_arr.append(right_arr[j])
                    j += 1
        return new_arr
```

```python
sort(arr)
```
